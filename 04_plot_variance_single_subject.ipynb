```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['svg.fonttype'] = 'none'
```

```python
emb = np.load('singles_embedding_dense_cerebellumonly_res.npy')
```

```python
lam = np.squeeze(np.array(list(emb.item()['lambdas'].flatten())))

fig, (ax) = plt.subplots(1, 2, figsize=(15,7), facecolor='white')
fontdictdetermine = {'family': 'helvetica',
        'color':  'black',
        'weight': 'normal',
        'size': 25,
        }
vals = np.cumsum(lam) / np.sum(lam)
vals = np.hstack((0,vals))
vals = np.hstack((0,vals))
vals = lam/ np.sum(lam)
ax[1].semilogx(vals, marker='d', markersize='15.0', mfc='k', mec='w', linestyle='dashdot', color='k')
ax[1].set_ylim([-0.005,np.max(vals)+0.005])
ax[1].set_xlim(xmin=.9)
ax[1].set_title(' ', fontdict = fontdictdetermine)

plt.yticks(size=15)
plt.xticks(size=1)

ax[1].set_xlabel('Gradient number', fontdict = fontdictdetermine)
ax[1].set_ylabel('Variance explained by gradient', fontdict = fontdictdetermine)
plt.tight_layout()

fig.savefig('figure.lambdas.svg', format='svg')
fig.savefig('figure.lambdas.png')
plt.show()
```
